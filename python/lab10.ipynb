## 1. Wprowadzenie do wizualizacji danych przy użyciu biblioteki matplotlib

```
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
```

### 1.1 Prosty wykres funkcji liniowej

```
plt.plot([1, 2, 3, 4])
plt.ylabel('jakieś liczby')
plt.show()
```

### 1.2 Różne style wykresów

#### 1.2.1 Wykres funkcji kwadratowej ze wskazaniem koloru i stylu linii oraz zakresem wartości na osiach x i y

```
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro-')
plt.axis([0, 6, 0, 20])
plt.show()
```

#### 1.2.2 Zestawienie dwóch wykresów funkcji kwadratowej o róznych stylach



```
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'o')

plt.axis([0, 6, 0, 20])
plt.show()
```

#### 1.2.3 Wygenerowanie wielu wykresów za pomocą pojedyńczego wywołania funkcji plot

```
t = np.arange(0., 5., 0.2)
```

```
t
```

```
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 1.2.4 Wygenerowanie wielu wykresów z legendą

```
x = np.linspace(0, 2, 100)
```

```
x[:10]
```

```
plt.plot(x, x, label='liniowa')
plt.plot(x, x**2, label='kwadratowa')
plt.plot(x, x**3, label='sześcienna')

plt.xlabel('etykieta x')
plt.ylabel('etykieta y')

plt.title("Prosty wykres")

plt.legend(loc='center')

plt.show()
```

#### 1.2.5 Wykres punktowy

```
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```
data
```

```
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('wartość a')
plt.ylabel('wartość b')
plt.show()
```

## 1.3 Wielowykresy

```
x1 = np.arange(0.0, 2.0, 0.02)
x2 = np.arange(0.0, 2.0, 0.02)

y1 = np.sin(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```
x1[:10]
```

```
y1[:10]
```

```
plt.subplot(2, 1, 1)
plt.plot(x1, y1, '-')
plt.title('Dwa podwykresy')
plt.ylabel('sin(x)')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r-')
plt.xlabel('x')
plt.ylabel('cos(x)')

plt.show()
```

### 1.3.2 Pomijanie umieszczania wykresu w różnych miejsach siatki

```
x1 = np.arange(0.0, 2.0, 0.02)
x2 = np.arange(0.0, 2.0, 0.02)
y1 = np.sin(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(3, 2, 1)
plt.plot(x1, y1, '-')
plt.title('Dwa podwykresy')
plt.ylabel('sin(x)')

plt.subplot(324)
plt.plot(x2, y2, 'r-')
plt.xlabel('x')
plt.ylabel('cos(x)')

plt.subplot(3, 2, 5)
plt.plot(x2, y2, 'r-')
plt.xlabel('x')
plt.ylabel('cos(x)')

plt.show()
```

## 1.4 Wykres słupkowy

```
etykiety = ['K', 'M']
wartosci = [345, 435]

plt.bar(etykiety, wartosci)

plt.xticks(rotation=45)
plt.ylabel('Ilość narodzin')
plt.xlabel('Płeć')
plt.show()
```

## 1.5 Histogram

```
x = np.random.randn(10000)
```

```
x[:5]
```

```
plt.hist(x, bins=50, facecolor='g', alpha=0.75, density=True)

plt.xlabel('Wartości')
plt.ylabel('Prawdopodobieństwo wystąpienia')
plt.title('Histogram')
plt.grid(True)
plt.show()
```

## 1.6 Wykres kołowy

```
zawodnicy = ['Messi', 'Suarez', 'Dembele', 'Coutinho']
bramki = [48, 25, 13, 11]

wedges, texts, autotexts = plt.pie(bramki, labels=zawodnicy, autopct=lambda pct: '{:.1f}%'.format(pct), textprops=dict(color='black'))
plt.setp(autotexts, size=10, weight='bold')
plt.title('Pierwsza wersja wykresu')
plt.legend(title='Zawodnicy', loc='upper left')
plt.show()
```
